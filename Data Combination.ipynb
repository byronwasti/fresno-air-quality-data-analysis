```python
%matplotlib inline

import numpy as np
import pandas as pd
import functools
```

```python
!ls data/Fresno/
```

```python
def read_file(filename):
    df = pd.read_csv(filename, parse_dates=[1])
    df.dropna(subset=['value'], inplace=True)
    df.date = pd.to_datetime(df.date, format="%Y-%m-%d-%H")
    df.index = df.date
    df.sort_index(inplace=True)
    return df
```

```python
df_ozone = read_file("data/Fresno/data_combination/OZONE.csv")
df_ozone.head()
```

```python
df_pm10 = read_file("data/Fresno/data_combination/PM10.csv")
df_pm10.head()
```

```python
df_pm25 = read_file("data/Fresno/data_combination/PM25.csv")
df_pm25.head()
```

```python
def split_into_list(x):
    if x == 'CALM':
        return (0, 0)
    else:
        y = x.split('/')
        return (float(y[0]), float(y[1]))

df_wind = read_file("data/Fresno/data_combination/WINSPD.csv")
df_wind['value_split'] = df_wind.value.apply(split_into_list)
df_wind['wd'] = df_wind.value_split.apply(lambda x: x[0])
df_wind['ws'] = df_wind.value_split.apply(lambda x: x[1])
df_wind.head()
```

```python
df_nox = read_file("data/Fresno/data_combination/NOX.csv")
df_nox.head()
```

```python
df_temp = read_file("data/Fresno/data_combination/TEMP.csv")
df_temp.head()
```

```python
df_merge = pd.merge(df_ozone, df_wind, how='outer', left_index=True, right_index=True, copy=False, suffixes=('_ozone', '_wind'))
df_merge2 = pd.merge(df_pm10, df_pm25, how='outer', left_index=True, right_index=True, copy=False, suffixes=('_pm10', '_pm25'))
```

```python
df_all = functools.reduce(lambda left,right: pd.merge(left, right, how='outer', left_index=True, right_index=True, copy=False), [df_ozone, df_wind, df_pm10, df_pm25, df_temp, df_nox])
df_all.head()
```

```python
df_all.to_csv('data_merged_all.csv')
```
